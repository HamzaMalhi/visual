```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/usamaali/Music/netflix_titles.csv")
df.head()
```

```python
country = df['country'].str.split(', ', expand=True).stack().str.replace('[^a-zA-Z\s]+', '',regex=True).value_counts().to_frame().reset_index()
country.columns = ['country','count']
country = country.groupby(['country'], as_index=False).sum()
```

```python
#using matplot
topcountry = country.sort_values(by='count', ascending=False).head(10)
#Plot the chart
plt.bar(topcountry['country'], topcountry['count'], color='#db0000')

#Set the title and axis labels
plt.title('Top 10 Countries of Origin for Content on Netflix', fontsize=20)
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend()
#Show the plot
plt.show()
```
