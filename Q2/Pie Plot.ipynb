```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/usamaali/Music/netflix_titles.csv")
df.head()
```

```python
# Using matplot
# using matplot
# Count the number of occurrences of each item
counts = df['rating'].value_counts()
#Plot the chart
plt.pie(counts.values, labels=counts.index, colors=['#db0000', '#000000', '#f10000','#666666' ,'#ffcccc', '#ff0000', '#cccccc', '#ff6666',
'#ff9999', '#ff3333', '#fc0000','#999999', '#e60000','#ffffff'])

plt.title("Movies vs TV Shows", fontsize=20)
plt.axis('equal')
plt.show()
```
